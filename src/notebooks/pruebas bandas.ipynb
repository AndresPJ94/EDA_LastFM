```python
import pandas as pd 
import numpy as np

#LIBRER√çAS DE MATPLOTLIB
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("seaborn-colorblind")


#LIBRERIAS DE PLOTLY
from plotly.offline import init_notebook_mode, iplot, plot
import plotly as py
init_notebook_mode(connected=True)
import plotly.graph_objs as go
from wordcloud import WordCloud

```

```python
df = pd.read_csv("200.txt")
df.columns=['banda', 'genero1', 'genero2', 'genero3', 'genero4', 'genero5', 'genero6']
a = df['genero1'].append(df['genero2']).reset_index(drop=True).append(df['genero3']).reset_index(drop=True).append(df['genero4']).reset_index(drop=True).append(df['genero5']).reset_index(drop=True).append(df['genero6']).reset_index(drop=True)
w = a.dropna().value_counts().head(10).to_frame()
w

def func(pct):
 return "{:1.1f}%".format(pct)
plot = w.plot.pie(subplots=True, legend=False, autopct=lambda pct: func(pct))
plt.axis('off')
plt.show()
```

```python
df = pd.read_csv("200.txt")
df = df.head(50)
df.columns=['banda', 'genero1', 'genero2', 'genero3', 'genero4', 'genero5', 'genero6']
a = df['genero1'].append(df['genero2']).reset_index(drop=True).append(df['genero3']).reset_index(drop=True).append(df['genero4']).reset_index(drop=True).append(df['genero5']).reset_index(drop=True).append(df['genero6']).reset_index(drop=True)
data = a.dropna().value_counts().head(10)
pd.DataFrame(data)

my_circle=plt.Circle( (0,0), 0.7, color='white')

plt.figure(figsize=(7,7))
plt.pie(data.values,
        labels=data.index,
        autopct='%1.2f%%')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
p.savefig('outputs/corona50.png')
```
